# Statistical Distributions


We are going to use numpy, Matplotlib and seaborn

NumPy library is used for numerical operations and generating random data. <br>
Matplotlib provides tools for creating static visualizations. <br>
Seaborn is built on Matplotlib to create more attractive and informative statistical graphics.

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
```

This code uses **Seaborn** to plot the probability density functions (PDFs) of three different distributions:

- **Normal distribution**: Generated with a mean (`loc`) of 50 and a standard deviation (`scale`) of 7, representing a symmetric, bell-shaped curve.
- **Poisson distribution**: Generated with a rate parameter (`lam`) of 50, representing the distribution of rare events happening over a fixed interval.
- **Binomial distribution**: Generated with 1000 trials (`n`) and a small probability of success (`p=0.01`), showing the distribution of binary outcomes across multiple experiments.

## 1. Normal Distribution

The **Normal Distribution** is a continuous probability distribution that is symmetrical around the mean. It is often referred to as the Gaussian distribution and follows the classic "bell curve". The distribution is completely defined by its mean (`μ`) and standard deviation (`σ`).

Key characteristics:
- Symmetrical and bell-shaped.
- Defined by two parameters: **mean** and **standard deviation**.
- Widely used in statistics and natural sciences.

Let's visualize a normal distribution with a mean of 50 and a standard deviation of 7.


```python
sns.kdeplot(random.normal(50, 7, 1000), label='Normal', fill=True)
plt.xlabel('Value'); plt.ylabel('Density'); plt.title('Normal Distribution')
plt.legend(); plt.show()
```

## 2. Binomial Distribution

The **Binomial Distribution** describes the number of successes in a fixed number of independent trials, where each trial has two possible outcomes: success or failure.

Key characteristics:
- Describes **discrete** events (success/failure).
- Defined by two parameters: **number of trials** (`n`) and **probability of success** (`p`).
- The shape of the distribution can be symmetric or skewed depending on the values of `n` and `p`.

Now, let's visualize a binomial distribution with 1000 trials and a success probability of 0.01.


```python
sns.kdeplot(random.binomial(1000, 0.01, 1000), label='Binomial', fill=True)
plt.xlabel('Value'); plt.ylabel('Density'); plt.title('Binomial Distribution')
plt.legend(); plt.show()
```

## 3. Poisson Distribution

The **Poisson Distribution** describes the probability of a given number of events occurring within a fixed interval of time or space, where these events happen independently and at a constant rate.

Key characteristics:
- Describes **discrete** events.
- Defined by one parameter: the **rate of occurrence** (`λ`).
- Typically used for rare events over time or space (e.g., the number of phone calls received in an hour).

Let's visualize a Poisson distribution with a rate parameter (`λ`) of 50.


```python
sns.kdeplot(random.poisson(50, 1000), label='Poisson', fill=True)
plt.xlabel('Value'); plt.ylabel('Density'); plt.title('Poisson Distribution')
plt.legend(); plt.show()
```

## 4. Comparison of Normal, Binomial, and Poisson Distributions

Now, let's visualize all three distributions on the same plot for a direct comparison. This will help us see how they differ in terms of shape and behavior.

We will plot:
- **Normal Distribution** with mean 50 and standard deviation 7.
- **Binomial Distribution** with 1000 trials and a success probability of 0.01.
- **Poisson Distribution** with a rate parameter (`λ`) of 50.


```python
sns.kdeplot(random.normal(50, 7, 1000), label='Normal', fill=True)
sns.kdeplot(random.binomial(1000, 0.01, 1000), label='Binomial', fill=True)
sns.kdeplot(random.poisson(50, 1000), label='Poisson', fill=True)
plt.xlabel('Value'); plt.ylabel('Density'); plt.title('Comparison of Distributions')
plt.legend(); plt.show()
```

## 5. Conclusion

We explored the **Normal**, **Binomial**, and **Poisson** distributions, visualizing their PDFs. Each distribution has distinct characteristics and applications in statistics and probability theory.
